## <center> 2 LASSO & Inpainting </center>

### **I) Introduction :**

L'**Inpainting** est une tâche qui consiste à synthétiser et à reconstruire des parties endommagées ou manquantes d'une image. Le principal challenge consiste à maintenir la sémantique globale de l'image et à générer une texture la plus réaliste possible pour les régions manquantes. Dans la litterature, il existe deux approches possible pour cette tâche : la première (PDE based : [[4]](https://dl.acm.org/doi/pdf/10.1145/344779.344972) ) qui consiste à remplir étape par étape une region manquante en commencant par ses contours jusqu'au pixel central de la region, la deuxième approche approxime le patch détérioré en utilisant le patch le plus proche (selon une certaine métrique) parmi ceux ne contenant pas de pixels manquants. Dans ce projet, on considérera deux heuristiques pour l'ordre de remplissage des patchs et l'algorithme du Lasso pour la prediction des pixels manquants. Ce dernier offre une solution "sparse" en terme de poids ce qui permet de considérer uniquement les atomes du dictionnaire qui sont réellement nécessaires pour la reconstruction du patch détérioré, ce qui aide entre autre à garder une cohérence en terme de texture et de couleurs. Autrement dit, le Lasso effectue une selection de features.

Il est important de préciser que l'ordre de remplissage est très important, en effet, la qualité de l'image reconstruite est fortement influencée par l'ordre de remplissage, un ordre de remplissage optimal devrait donner une forte priorité aux patchs qui se situent dans des regions continues en terme de structure et de texture ou des patchs contenant peu de pixels manquants par exemple.

On etudiera notamment l'effet de l'hyper-paramètre alpha sur le résultat mais aussi des autres hyper-paramètres du problème, à savoir :  la taille de la région détériorée, le pas de construction des patchs et leur taille.

#### Fonctions utiles

```python
CONST = -100

def read_im(file):
    """
    Fonction permettant de lire un fichier image.
    -------------------------------------------
    Args:
        - file : chemin vers le fichier
    """
    img = plt.imread(file)
    if np.max(img) > 200:
        img = img/255.
    return rgb_to_hsv(img) - 1

def display_im(img,h=20):
    """
    Affichage d'une image.
    -------------------------------------------
    Args:
        - img : image à afficher.
    """
    img = img.copy()
    if len(img.shape) == 1 or img.shape[1] == 1:
        img = vec_to_patch(img,h)
    img[img <= CONST] = -1
    plt.imshow(hsv_to_rgb(img+1))
    
def show(img_original, img_damaged, img_restored,titles=None,h=20):
    """
    Permet un affichage des trois images : l'image original, 
    l'image deteriorée et l'image reconstruite.
    -------------------------------------------
    Args:
        - img_original : image original.
        - img_damaged : image détériorée.
        - img_restored : image reconstruire.
    """
    if titles is None:
        titles = ['Original image','Damaged image','Restored image']
    plt.figure(figsize= (20, 10))
    plt.subplot(131)
    display_im(img_original,h)
    plt.title(titles[0])

    plt.subplot(132)
    display_im(img_damaged,h)
    plt.title(titles[1])
    plt.subplot(133)
    display_im(img_restored,h)
    plt.title(titles[2])

    plt.show()    
def get_patch(img,i,j,h):
    """
    Retourne un patch de l'image centré en (i,j) et de longueur h.
    -------------------------------------------
    Args:
        - img : image considérée.
        - (i,j) : coordonnées du centre du patch.
        - h : longueur du patch.
    """
    return img[ i - h//2 : i + h//2, j - h//2 : j + h//2, : ].copy()

def patch_to_vec(patch):
    """
    Transformation d'un patch en vecteur.
    -------------------------------------------
    Args:
        - patch : patch considéré.
    """
    return patch.reshape(-1)

def vec_to_patch(vector,h):
    """
    Transformation d'un vecteur en patch.
    -------------------------------------------
    Args:
        - vector : vecteur considéré.
        - h : longueur du patch.
    """
    return vector.reshape((h,h,3))

def noise(img, prc):
    """
    Supprime des pixels aléatoirement dans une image.
    -------------------------------------------
    Args:
        - img : image considérée.
        - prc : pourcentage de pixels à supprimer.
    """
    noisy_img = img.copy().reshape((-1,3))
    height, width = img.shape[:2]
    n = int(prc*height*width)
    noisy_img[np.random.randint(0,height*width,n),:] = CONST
    return noisy_img.reshape(height,width,3)
   

def delete_rect(img, i, j, height, width):
    """
    Supprime un patch d'une image
    -------------------------------------------
    Args:
        - img : image considérée.
        - (i,j) : coordonnées du centre du patch.
        - height : longueur du patch.
        - width : largeur du patch.
    """
    img = np.copy(img)
    img[(i-height//2):(i+height//2), (j-width//2):(j+width//2)] = CONST
    return img, get_patch(img,i,j,height)

def valid_patch(img,i,j,h=20):
    """
    Teste si un patch est valide dans une image 
    -------------------------------------------
    Args:
        - img : image considérée.
        - (i,j) : coordonnées du centre du patch.
        - h : longueur du patch.
    """
    return i - h//2 >= 0 and j - h//2 >= 0 and i + h//2  <= img.shape[0] and j + h//2 <= img.shape[1]

def get_dictionary(img, h, step=20):
    """
    Retourne le dictionnaire d'atomes et les patchs contenant des pixels manquants
    -------------------------------------------
    Args:
        - img : image considérée.
        - h : longueur du patch
        - step : le pas considérée pour construire les patchs.
    """
    missing = []
    complete = []
    for i in range(0, img.shape[0], step):
          for j in range(0, img.shape[1], step):
                if valid_patch(img, i, j, h):
                    patch = get_patch(img,i,j,h)  
                    if (np.sum(patch[:,:,0] <= CONST) == 0):
                        complete.append(patch_to_vec(patch))
                    else:
                        missing.append(patch_to_vec(patch))

    return np.asarray(complete),np.asarray(missing)


def filled_pixel(img, i, j):
    """
    Teste si le pixel est manquant ou pas.
    -------------------------------------------
    Args:
        - img : image considérée.
        - (i,j) : coordonnées du pixel.
    """
    return img[i,j,0] != CONST

def approx(patch, dic, alph, max_iter=5000):
    """
    Retourne le vecteur de poids sur le dictionnaire qui approxime au mieux un patch.
    -------------------------------------------
    Args:
        - patch : patch considéré.
        - dic : dictionnaire d'atomes.
        - alph : hyper-paramètre alpha du Lasso.
        - max_iter : nombre d'itérations du Lasso.
    """
    patch_ = patch.copy()
    clf = Lasso(alpha=alph,max_iter=max_iter)
    patch_ = patch_.ravel()
    dic = dic.T
    train_ind = np.argwhere(patch_ != -100).ravel()
    clf.fit( dic[train_ind], patch_[train_ind])
    test_ind = np.argwhere(patch_ == -100).ravel()
    if len(test_ind) > 0:
        patch_[test_ind] = clf.predict(dic[test_ind])
    
    return patch_, clf.coef_

tree_small = read_im('tree-small.jpg')
bordeo = read_im('borneo.jpg')
wall = read_im('brick-small.jpg')
tree_big = read_im('tree.jpg')
```

#### Nous pouvons déduire à partir des expériences ci-dessus les deux points suivants:
1.   **La taille du dictionnaire** : il est evident que lorsque le dictionnaire est construit à partir d'une image dont la majorité des pixels ne sont pas manquants, nous avons à notre disposition plusieurs signaux elementaires permettant de reconstruire au mieux le patch détérioré. En revanche, lorsque le dictionnaire est construit à partir d'une image bruité, le dictionnaire d'atomes est très petit, par conséquent, la combinaison linéaire approxime très mal voir pas du tout le patch détérioré. On peut néanmoins diminuer la taille des patchs afin d'augmenter la taille du dictionnaire.

2.   **Influence de l'hyper-paramètre alpha** : Nous remarquons d'abord que le nombre de coefficients non nuls dépend fortement de la valeur de alpha, en effet, le nombre de variable sélectionnées par le Lasso diminue lorsque alpha devient grand, c’est-à-dire, si alpha est grand, la contrainte exercé sur le vecteur de poids l’est également, et inversement. Cela peut notamment se voir sur les valeurs de la norme. En d'autres termes, alpha controle le compromis entre l'erreur de reconstruction et la parcimonie de la solution, on peut observer ci-dessus que lorsque alpha se rapproche de 0, le modèle prend en compte un nombre considérable d'atomes et approxime très bien le patch détérioré.  

## <center> Implémentation de l'inpainting et heuristiques de remplissage </center>

 **Première heuristique** : Dans un premier temps, on teste une heuristique naive qui consiste à considérer d'abord les patchs les moins détériorés (ie contenant le moins de pixels manquants), on remplit au fur et à mesure en commencant par les bords jusqu'au centre de l'image.

```python

class Inpainting:
    """
    Implémentation de l'inpainting.
    -----------------------------------------------------------
    Parameters : 
        - img : image considérée.
        - h : longueur des patchs.
        - (point_x, point_y) : coordonnées du pixel central de la region détériorée. 
        - (width, height) : dimensions de la région détériorée.
        - step : pas considéré pour la construction des patchs
        - alpha : hyper-paramètre du Lasso.
        - max_iter : Nombre d'itérations du Lasso.
        - perc_noise : pourcentage de pixels à supprimer lorsque la tache consiste à débruiter un patch.
        - noisy : si True, on ajoute du bruit au patch.
    """
    def __init__(self, img, h,point_x, point_y, width, height,step=20, alpha=0.01, max_iter=50000,perc_noise =.3, noisy=False):
        self.alpha = alpha
        self.max_iter = max_iter
        self.h = h
        self.img_original = img.copy()
        if noisy : 
            self.image = img.copy()
            patch = get_patch(img,point_x,point_y,width)
            self.image[point_x-width//2:point_x+width//2,point_y-width//2:point_y+width//2,:] = noise(patch, perc_noise)
        else:
            self.image = delete_rect(img, point_x, point_y,height, width)[0]
        self.source_region, self.target_regions = get_dictionary(self.image,self.h,step)
        self.source_region = self.source_region.T
        self.clf = Lasso(alpha=self.alpha, max_iter=self.max_iter)
        self.img_damaged = self.image.copy()
        
    def get_boundary(self):
        """
        Retourne les pixels centraux des patchs situés au bord de la région manquante.
        ---------------------------------------------------------------------------------------
        """
        pixels = []
        for i in range(self.image.shape[0]):
            for j in range(self.image.shape[1]):
                if not filled_pixel(self.image, i,j):
                    patch = get_patch(self.image,i,j,self.h)
                    #si c'est un patch situé sur le contour de l'objet retiré
                    if np.sum(patch[:,:,0] == CONST) < (self.h * self.h):
                        pixels.append((i,j))
        return pixels
    def get_next_patch(self, boundary):
        """
        Retourne le patch contenant le moins de pixels manquants parmis ceux situés au bord de la région manquante.
        -------------------------------------------------------------------------------------------
        """
        ind = np.argmax([np.count_nonzero(get_patch(self.image,p[0],p[1],10)[:,:,0] != CONST) for p in boundary])
        pixel = boundary[ind]
        return pixel, get_patch(self.image,pixel[0],pixel[1],self.h)
            
    def fill_patch(self, i, j, values):
        """
        Remplit les pixels manquants par les valeurs prédites.
        -------------------------------------------------------------
        """
        self.image[ i- self.h//2 : i + self.h//2, j - self.h//2 : j + self.h//2, : ] = values.reshape(self.h,self.h,3)
    
    def fit_predict(self, display = True):
        """
        Fonction permettant l'apprentissage et la prédiction des pixels manquants.
        -------------------------------------------------------------------
        """
        init = time.time()
        boundary = self.get_boundary()
        while (len(boundary) > 0):
            gc.collect()
            curr_pixel, curr_patch = self.get_next_patch(boundary)
            curr_patch = curr_patch.ravel()
            train_ind = np.argwhere(curr_patch != -100).ravel()
            test_ind = np.argwhere(curr_patch == -100).ravel()
            self.clf.fit(self.source_region[train_ind], curr_patch[train_ind])
            curr_patch[test_ind] = self.clf.predict(self.source_region[test_ind])
            self.fill_patch(curr_pixel[0], curr_pixel[1], curr_patch)
            boundary = self.get_boundary()
        print(" time: {0:.2f} seconds".format(time.time() - init))
        self.img_restored = self.image.copy()
        if display: 
            show(self.img_original,self.img_damaged,self.img_restored)
```

```python
inpaint = Inpainting(tree_small, 10,40, 20, 20, 20,step=2, alpha=0.001, max_iter=5000,perc_noise =.2, noisy=True)
inpaint.fit_predict()
```

### Observations : 

Nous constatons que alpha contribue de manière significative au résultat final. lorsque alpha est grand, nous restreignons le dictionnaire à peu de patchs donc nous faisons l'hypothèse que seuls quelques atomes sont nécessaires pour la reconstruction, on peut dire qu'on est clairement dans un cas de sous-apprentissage. Inversement, si alpha est petit, nous considérons beaucoup plus de patchs et donc plus de couleurs et de textures lors de la reconstruction.


Lorsque alpha vaut 0, le Lasso correspond à une régression linéaire classique, donc pas de selection de features, l'algorithme prend en compte tous les patchs du dictionnaire ce qui entraine une incohérence dans les couleurs et la texture de la région reconstruite. 

## Deuxième heuristique : Implementation de l'article [[3]](https://www.irisa.fr/vista/Papers/2004_ip_criminisi.pdf)

### Formalisation : 
 On considère une image $\textit{I}$ qui contient une region manquante qu'on souhaite reconstruire et qu'on appelera $\Omega$, on définit le dictionnaire d'atomes $\phi$ composé de patchs ne contenant aucun pixel manquant $\phi = \textit{I} - \Omega$.

 Pour chaque pixel, on définit une mesure qu'on nommera "confidence" $C$ qui mesure à quel point on est sur que la valeur du pixel est effectivement la bonne. Tout au long de l'algorithme, chaque patch se voit attribuer une priorité  qui détermine l'ordre de remplissage.

 **Calcul des priorités** :  l'heuristique consiste à favoriser les patchs qui se situe sur des contours epais et ayant autour des pixels avec une forte valeur de C.

 Pour un patch $\psi_p$ centré sur un pixel $p$, on définit sa priorité de la manière suivante : 
### $$P(p) = C(p) D(p)$$

 avec $C(p) = \frac{\sum_{q \in \psi_p \cap \phi} C(q)}{|\psi_p|}$ et $D(p) = \frac{|\nabla I_p^\perp n_p|}{|\alpha|}$

 |$\psi_p$| est l'aire du patch considéré, $n_p$ est un vecteur unitaire orthogonal au contour de l'objet manquant au point $p$, $\nabla I_p^\perp$ est l'isophote au point $p$ qui mesure l'intensité et la direction du changement de luminosité.

 **Mise à jour de C** : Après le remplissage du patch $\psi_p$ avec les valeurs predites par le Lasso, on met à jour le terme C pour chaque pixel $q \in \psi_p \cap \Omega$ (ie les pixels appartenant au patch et à la region manquant, autrement dit, ceux qu'on reconstruit) : 

### $$C(q) = C(p)$$

### Pseudo Code : 
### -     Repeter jusqu'à ce que tous les pixels soient remplis : 
1.    Extraire le contour de l'objet manquant $\delta \Omega$ à l'aide du filtre de Sobel
2.    Calculer la priorité P de chaque pixel appartenant à $\delta \Omega$
3.    Prendre le patch $\psi_p$ tel que $p = argmax_{p \in \delta \Omega}P(p)$
4.    En utilisant le Lasso et à l'aide d'une combinaison linéaire des atomes $\phi$, reconstruire le patch précédent 
5.    Mettre à jour l'image
6.    Mettre à jour le terme C de chaque pixel rempli


```python
laplacian = np.array([[ 1.,  1.,  1.],[ 1., -8.,  1.],[ 1.,  1.,  1.]])
kerx = np.array([[ -1.,  0.,  1.], [-2.,  0.,  2.], [ -1.,  0.,  1.]])
kery = np.array([[ -1., -2.,  -1.], [ 0.,  0.,  0.], [ 1.,  2.,  1.]])
def get_front(masque):
    """
    Renvoie la frontière de l'objet manquant et les normes en chaque point.
    -------------------------------------------
    Args:
        - masque : masque de l'objet manquant 
        
    """
    dom, norm = [], []
    source  = 1 - masque
    lap = convolve2d(masque, laplacian,mode="same")
    gradX = convolve2d(source, kerx,mode="same")
    gradY = convolve2d(source, kery,mode="same")
    for i in range(lap.shape[0]):
        for j in range(lap.shape[1]):
            if lap[i,j] > 0:
                dom.append((i,j))
                dx,dy = gradX[i,j],gradY[i,j]
                n = np.sqrt(dy**2 + dx**2)
                if n != 0:
                    norm.append((dy/n,-dx/n))
                else : 
                    norm.append((dy,-dx))
    return np.asarray(dom), np.asarray(norm)
def extreme_points(img, h, point):
    """
    Renvoie les points extreme d'un patch 
    -------------------------------------------
    Args:
        - img : image considérée 
        - h : longueur du patch
        -point : centre du patch
    """
    px, py = point
    xsize, ysize, c = img.shape
    x_min, y_min = max(px - h//2, 0), max(py - h//2, 0)
    x_max, y_max = min(px + h//2, xsize - 1), min(py + h//2, ysize - 1)
    
    return((x_min, x_max),(y_min, y_max))

def compute_confidence(img, confidence, h, masque, domega): #confidence
    """
    Calcule le terme C (confidence) pour chaque point se situant sur le contour de l'objet manquant.
    -------------------------------------------
    Args:
        - img : image considérée.
        - confidence : terme C.
        - h : longueur du patch.
        - masque : masque de l'objet manquant.
        - domega : frontière de l'objet manquant.
    """
    for i in range(len(domega)):
        px, py = domega[i]
        points = extreme_points(img, h, domega[i])
        patch =get_patch(img,px,py,h)
        (x_min, x_max),(y_min, y_max) = points
        patch_size = h*h
        compteur = 0
        for x in range(x_min, x_max + 1):
            for y in range(y_min, y_max + 1):
                if masque[x, y] == 0: 
                    compteur += confidence[x, y]
        confidence[px, py] = compteur / patch_size
    return(confidence)

def compute_data(domega,norm,data,gradX,gradY):
    """
    Calcule le terme D (data) pour chaque point se situant sur le contour de l'objet manquant.
    -------------------------------------------
    Args: 
        - domega : frontière de l'objet manquant.
        - norm : norme en chaque point du contour.
        - data : terme D.
        - gradX : gradient dans la direction des x. 
        - gradY : gradient dans la direction des y.
    """
    for i in range(len(domega)):
        x, y = domega[i]
        nx, ny = norm[i]
        data[x,y] = np.sqrt((gradX[x, y] * nx)**2 + (gradY[x, y] * ny)**2) / 255.
    return data
    
def compute_priority(img, h, masque, domega, norm, data, gradX, gradY, confidence):
    """
    Calcule la priorité de chaque patch.
    -------------------------------------------
    Args:
        - img : image considérée.
        - h : longueur des patchs.
        - masque : masque de l'objet manquant.
        - domega : contour de l'objet manquant.
        - norm : norme en chaque point du contour.
        - data : 
        - gradX : gradient dans la direction des x.
        - gradY : gradient dans la direction des y.
    """
    C = compute_confidence(img, confidence, h, masque, domega)
    D = compute_data(domega, norm, data, gradX, gradY)
    index = np.argmax([C[i,j] * D[i,j] for (i,j) in zip(domega[:,0],domega[:,1])])
    return (C, D, index)

def fill_patch(img,h, i, j, values,masque,confidence):
    """
    Remplit les pixels manquants par les valeurs prédites et met à jour la confidence.
    -------------------------------------------
    Args:
        - img : image considérée.
        - h : longueur du patch.
        - (i,j) : centre du patch.
        - values : valeurs prédites.
        - masque : masque de l'objet manquant.
        - confidence : terme C
    """
    img[ i- h//2 : i + h//2, j - h//2 : j + h//2, : ] = values.reshape(h,h,3)
    masque[i- h//2 : i + h//2, j - h//2 : j + h//2] = 0
    confidence[i- h//2 : i + h//2, j - h//2 : j + h//2] = confidence[i,j]
    return img, masque, confidence


class Smart_Inpainting:
    """
    Implementation de l'inpainting en considérant l'heuristique décrite ci-dessus.
    -------------------------------------------
    Parameters:
        - img : image considérée.
        - h : longueur des patchs.
        - (point_x, point_y) : coordonnées du pixel central de la region détériorée. 
        - (width, height) : dimensions de la région détériorée.
        - step : pas considéré pour la construction des patchs
        - alpha : hyper-paramètre du Lasso.
        - max_iter : Nombre d'itérations du Lasso.
        - perc_noise : pourcentage de pixels à supprimer lorsque la tache consiste à débruiter un patch.
        - noisy : si True, on ajoute du bruit au patch.
    """
    def __init__(self, img, h, point_x, point_y, width, height, step=20, alpha=0.01, max_iter=50000, perc_noise =.3, noisy=False):
        self.alpha = alpha
        self.max_iter = max_iter
        self.h = h
        self.img_original = img.copy()
        if noisy : 
            self.image = img.copy()
            patch = get_patch(img,point_x,point_y,width)
            self.image[point_x-width//2:point_x+width//2,point_y-width//2:point_y+width//2,:] = noise(patch, perc_noise)
        else:
            self.image = delete_rect(img, point_x, point_y,height, width)[0]
        self.source_region, self.target_regions = get_dictionary(self.image,self.h,step)
        self.source_region = self.source_region.T
        self.clf = Lasso(alpha=self.alpha, max_iter=self.max_iter)
        self.img_damaged = self.image.copy()
        
        width, height, channels = self.image.shape
        self.masque = np.zeros((width,height))
        self.confidence = np.ones((width,height))
        self.data = np.ndarray(shape = self.image.shape[:2])
        
        damaged = np.where(self.image[:,:,0] == CONST)
        self.masque[damaged] = 1
        self.confidence[damaged] = 0
        self.omega = np.vstack([damaged]).T
        self.original = self.confidence.copy()
    
    def get_gradient(self):
        """
        Renvoie le gradient de l'image dans les directions x et y.
        -------------------------------------------
        
        """
        # hsv to grayscale
        gray_scale = cv2.cvtColor(self.image, cv2.COLOR_HSV2RGB)
        gray_scale = cv2.cvtColor(gray_scale, cv2.COLOR_RGB2GRAY)
        # calcul du gradient 
        gradientX = np.float32(cv2.convertScaleAbs(cv2.Scharr(gray_scale, cv2.CV_32F, 1, 0)))
        gradientY = np.float32(cv2.convertScaleAbs(cv2.Scharr(gray_scale, cv2.CV_32F, 0, 1)))
        gradientX[np.where(self.masque == 1)] = 0
        gradientY[np.where(self.masque == 1)] = 0
                    
        return gradientX/255., gradientY/255.
    
    def fit_predict(self,display = True):
        """
        Fonction permettant l'apprentissage et la prédiction des pixels manquants.
        -------------------------------------------
        
        """
        self.image = np.float32(self.image)
        gradientX, gradientY = self.get_gradient()
        keep_going = True
        init = time.time()
        while keep_going:
            gc.collect()
            gradientX, gradientY = self.get_gradient()
            domega, norm = get_front(self.masque)
            self.confidence, self.data, index = compute_priority(self.image,self.h,self.masque,domega,norm,self.data,gradientX, gradientY,self.confidence)
            curr_pixel = domega[index]
            curr_patch = get_patch(self.image,curr_pixel[0], curr_pixel[1],self.h)
            curr_patch = curr_patch.ravel()
            train_ind = np.argwhere(curr_patch != -100).ravel()
            test_ind = np.argwhere(curr_patch == -100).ravel()
            self.clf.fit(self.source_region[train_ind], curr_patch[train_ind])
            curr_patch[test_ind] = self.clf.predict(self.source_region[test_ind])
            self.image, self.masque, self.confidence = fill_patch(self.image,self.h,curr_pixel[0], curr_pixel[1], curr_patch, self.masque, self.confidence)
            if np.count_nonzero(self.masque) == 0:
                keep_going = False
                
        print(" time: {0:.2f} seconds".format(time.time() - init))
        self.img_restored = self.image.copy()
        if display:
            show(self.img_original,self.img_damaged,self.img_restored)
```

### **Remarques générales** : 
 1.     **Debruitage** : On remarque que la tache de débruitage est bien plus facile que la reconstruction de toute une région, cependant, comme il a été mentionné dans l'article [[1]](https://ieeexplore.ieee.org/document/4959679), l'inpainting ne consiste pas à faire du débruitage, si l'image est bruité, il serait impossible d'avoir un dictionnaire d'atomes permettant de reconstruire l'image, on a pu remarquer cela dans le série d'expériences que nous avons effectuée. Il faudra pour cela utiliser une autre méthode pour la construction du dictionnaire [[5]](https://ieeexplore.ieee.org/document/4392496).

 2.     **Influence de la taille des patchs et du pas de construction** : Après avoir essayer plusieurs valeurs pour la taille des patchs $h$ et du pas de construction $step$, nous avons conclut que ces deux hyper-paramètres dependent en realité de l'image et des objets contenus dans cette dernière. Pour une image plutôt unie en terme de structure et de texture, on peut considérer un pas plus grand, cependant, pour une image contenant beaucoup de détails, il faudra considérer un pas plus petit pour avoir assez de redondance dans le dictionnaire afin que l'algorithme soit capable de preserver les détails de l'image.

 3.       **Comparaison de l'algorithme avec l'état de l'art en inpainting** : La plupart des algorithmes de l'etat de l'art en inpainting utilise une méthode gloutonne pour la reconstruction qui consiste à selectionner un unique meilleur patch pour la reconstruction d'un patch ayant des pixels manquants, cela peut introduire une incohérence dans la reconstruction (notamment des objets indésirables), en revanche, l'algorithme présenté dans ce projet permet de pallier ce problème en faisant une reconstruction à l'aide d'une combinaison linéaire de patchs selectionnés au préalable par le Lasso.

4.       **Heuristique de remplissage** : La deuxième heuristique que nous avons implémenté (celle présentée dans l'article [[3]](https://www.irisa.fr/vista/Papers/2004_ip_criminisi.pdf) ) donne des résultats très satisfaisants car en considérant les isophotes de l'image, la méthode est capable de propager des structures linéaires ainsi que les textures vers la region détériorée. Cette heuristique est également très robuste sur les contours en préservant notamment leurs formes.

À travers ce projet, nous avons implémenté une technique d'inpainting présenté dans [[1]](https://ieeexplore.ieee.org/document/4959679) en considérant l'ordre de remplissage présenté dans [[3]](https://www.irisa.fr/vista/Papers/2004_ip_criminisi.pdf). Nous avons pu notamment mener des experiences sur le Lasso et sa capacité à faire un bon compromis entre la selection de features et l'erreur de reconstruction. Cet algorithme est néanmoins très couteux en temps de calcul sur des images haute résolution, plusieurs méthodes modernes beaucoup moins couteuses et beaucoup plus consistantes ont été developpé depuis, en utilisant notamment des réseaux de neurones à convolution (CNN) ou des réseaux adverses génératifs (GANs).
